```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# create data 
#create list of names of students
name = ['omar','samy','sally','hany','sameh',
'ahmed','mahmoud','magdy','osama','maged',
'john','joy','amr','bishoy','gamil',
'ibrahim','nabil','mina','tony','fady',
'david','Hanna','mostafa','joeshwoa','rina',
'hanan','salma','maria','carol','obdo',
'hamed','raouf','bola','samir','maher',
'tamer','reda','heba','noor','malak']

initial_id = 2020000
departments = ['AI','CS','IT','IS','DS']
num_of_students = 400
name_of_students = []
id_of_students = []
gpa_of_students = []
age_of_students = []
num_of_courses_for_each_student= []
department_of_students = []
for i in range(num_of_students):
    initial_id = initial_id +1
    name_of_students.append(name[np.random.randint(len(name))])
    id_of_students.append(initial_id)
    gpa_of_students.append(np.round(np.random.uniform(1,4),2))
    department_of_students.append(departments[np.random.randint(5)])
    num_of_courses_for_each_student.append(np.random.randint(15,35))
    age_of_students.append(np.random.randint(18,28))
```

```python
data = {'Name':name_of_students,'Id':id_of_students,'Gpa':gpa_of_students,
'Department':department_of_students,'Age':age_of_students,'Number Of Courses':num_of_courses_for_each_student}
df = pd.DataFrame(data,columns=data.keys())
print('Some of students are : ')
print(df.head(15)) # print 15 students of data frame 
print('--------------------------------------------------------------------')
print(df.describe()) # print statistics information about data
print('--------------------------------------------------------------------')
print(df.info())  # print information about data
print('--------------------------------------------------------------------')


```

```python
df = df.sort_values(['Gpa'],ascending=False) # sorting the data frame according to GPA attribute
##print(df)
df_ai = df.query('Department=="AI"').reset_index(drop=True) # get students whose department AI and store them in another fata frame , reset_index for intialize index starts with 0 , not old index that in the main data frame 
df_cs = df.query('Department=="CS"').reset_index(drop=True) # get students whose department CS and store them in another fata frame , reset_index for intialize index starts with 0 , not old index that in the main data frame 
df_is = df.query('Department=="IS"').reset_index(drop=True) # get students whose department IS and store them in another fata frame , reset_index for intialize index starts with 0 , not old index that in the main data frame 
df_it = df.query('Department=="IT"').reset_index(drop=True) # get students whose department IT and store them in another fata frame , reset_index for intialize index starts with 0 , not old index that in the main data frame 
df_ds = df.query('Department=="DS"').reset_index(drop=True) # get students whose department DS and store them in another fata frame , reset_index for intialize index starts with 0 , not old index that in the main data frame 
# print top 10 
print('TOP 10 For all departments are : ')
print(df.head(10).reset_index(drop=True))
print('--------------------------------------------------------------------')
# for visualize data
plt.figure(figsize=(7,6))
plt.scatter(df_ai['Age'],df_ai['Gpa'],marker='*',label='AI')
plt.scatter(df_is['Age'],df_is['Gpa'],marker='*',label='IS')
plt.scatter(df_cs['Age'],df_cs['Gpa'],marker='*',label='CS')
plt.scatter(df_ds['Age'],df_ds['Gpa'],marker='*',label='IT')
plt.scatter(df_ds['Age'],df_ds['Gpa'],marker='*',label='DS')
plt.xlabel('Age')
plt.ylabel('Gpa')
plt.title('Plotting Data using scatter plot graph for ALL Students Departments.')
plt.legend(loc='upper right')
plt.show()

plt.figure(figsize=(11,9))
plt.pie([df_ai['Gpa'].count(),df_cs['Gpa'].count(),df_is['Gpa'].count(),df_it['Gpa'].count(),df_ds['Gpa'].count()],labels=['AI Students','CS Students','IS Students','IT Students','DS Students'],autopct='%1.2f%%',shadow=True,explode=[.03,.03,.03,.03,.03])
plt.legend(loc='upper right')
plt.title('Plotting Number of students that in all Departments.')
plt.show()

#plotting with 3d
fig=plt.figure(figsize=(16,9))
ax = plt.axes(projection='3d')
ax.grid(b=True,color='gray',linstyle='-.',linewidth=0.3,alpha=0.2)
cmap_ = plt.get_cmap('hsv')
s=ax.scatter3D(df['Age'],df['Gpa'],df['Number Of Courses'],c=(df['Age']+df['Gpa']+df['Number Of Courses']),cmap=cmap_)
ax.set_xlabel('Age',fontweight='bold')
ax.set_ylabel('Gpa',fontweight='bold')
ax.set_zlabel('Number Of Courses',fontweight='bold')
plt.title('3D Scatter plot for all students that in all departments ')
fig.colorbar(s,ax=ax,shrink=0.5,aspect=5)
plt.show()

fig2=plt.figure()
ax2 = plt.axes(projection='3d')

s2=ax2.scatter3D(df_ai['Age'],df_ai['Gpa'],df_ai['Number Of Courses'],label='AI',marker='+')
s3=ax2.scatter3D(df_cs['Age'],df_cs['Gpa'],df_cs['Number Of Courses'],label='CS',marker='+')
s4=ax2.scatter3D(df_is['Age'],df_is['Gpa'],df_is['Number Of Courses'],label='IS',marker='+')
s5=ax2.scatter3D(df_it['Age'],df_it['Gpa'],df_it['Number Of Courses'],label='IT',marker='+')
s6=ax2.scatter3D(df_ds['Age'],df_ds['Gpa'],df_ds['Number Of Courses'],label='DS',marker='+')
ax2.set_xlabel('Age',fontweight='bold')
ax2.set_ylabel('Gpa',fontweight='bold')
ax2.set_zlabel('Number Of Courses',fontweight='bold')
plt.title('3D Scatter plot for all students that in all departments ')
plt.legend(loc='upper right')
plt.show()

plt.boxplot(df['Gpa'])
plt.title('Plotting Data using box plot graph for All Students Department.')
plt.show()
```
